```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import nflfastpy as nfl
```

```python
yearly_url = 'https://raw.githubusercontent.com/fantasydatapros/data/master/yearly/{year}.csv'
start_year = 2013
end_year = 2020 # As of 6/22/22, the repository above only contains yearly data up to and including the 2020 season. 

yearly_df = pd.DataFrame()
for year in range(start_year, end_year+1):
    year_df = pd.read_csv(yearly_url.format(year=year))
    year_df['Season'] = year
    col_name = year_df.columns.tolist()
    col_name.pop()
    col_name.insert(0, 'Season')
    year_df = year_df[col_name]
    yearly_df = pd.concat([yearly_df, year_df])
yearly_df.drop(columns=['Unnamed: 0'], inplace=True)
print(yearly_df.shape)
yearly_df.head()
```

```python
# Attach scraped 2021 yearly season data to yearly_df 

year_2021_df = pd.read_csv('/Users/brandynklee/Python_Projects/fantasy_footballl_22/data/yearly_2021.csv')
year_2021_df['Season'] = 2021
year_2021_df = year_2021_df[col_name]

yearly_df = pd.concat([yearly_df, year_2021_df])
print(yearly_df.shape)
yearly_df.head()
```
