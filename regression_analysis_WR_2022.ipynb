```python
# Grab roster data from nflfastr

roster_url = 'https://raw.githubusercontent.com/nflverse/nflfastR-roster/master/data/seasons/roster_{year}.csv'
s_year = 2013
e_year = 2021
roster_df = pd.DataFrame()

for year in range(s_year, e_year+1):
     r_df = pd.read_csv(roster_url.format(year=year))
     r_df['Season'] = year
     col_name = r_df.columns.tolist()
     col_name.pop()
     col_name.insert(0, 'Season')
     r_df = r_df[col_name]
     roster_df = pd.concat([roster_df, r_df])
roster_df.drop(columns=['season'], inplace=True)
print(roster_df.shape)
roster_df.head()
```

```python
# Load snapcount data for seasons of 2013-2021 inclusive from scraped csv file.

snapcount_df = pd.read_csv('/Users/brandynklee/Python_Projects/fantasy_footballl_22/data/snapcount_compiled_2013_2021.csv')
snapcount_df.drop(columns=['Unnamed: 0'], inplace=True)
print(snapcount_df.shape)
snapcount_df.head()
```

```python
# As today's focus is on WR, we will be looking at air yards via play by play data from nflfastr (https://github.com/nflverse/nflfastR).
# Such data will be used to calculate the total amount of air yards per season per player. 

pbp_df = pd.DataFrame()
y1 = 2013
y2 = 2021

for year in range(y1, y2+1):
    yearly_pbp_df = nfl.load_pbp_data(year)

    yearly_pbp_df = yearly_pbp_df.loc[yearly_pbp_df['pass_attempt'] == 1] # Because we want to look at air yards, we only look at plays where a pass was attempted
    # yearly_pbp_df = yearly_pbp_df.groupby('receiver_id', as_index=False)['air_yards'].sum()
    yearly_pbp_df = yearly_pbp_df.groupby(['receiver_id']).agg({'air_yards':'sum', 'yards_after_catch':'sum'}).reset_index()
    yearly_pbp_df['Season'] = year
    col_name = yearly_pbp_df.columns.tolist()
    col_name.pop()
    col_name.insert(0, 'Season')
    yearly_pbp_df = yearly_pbp_df[col_name]
    pbp_df = pd.concat([pbp_df, yearly_pbp_df])

print(pbp_df.shape)
pbp_df.head()
```

```python
wr_analysis_df = snapcount_wr_df.merge(
    yearly_wr_df, on=['Player', 'Season'], how='left'
)\
.merge(
    roster_wr_df, on=['Player','Season'], how='left'
)\
.merge(
    pbp_df.rename({'receiver_id':'gsis_id'}, axis =1)[['gsis_id','Season', 'air_yards', 'yards_after_catch']], on=['gsis_id', 'Season'], how='left'
)

wr_analysis_df.head()
```

```python
wr_analysis_df = wr_analysis_df[[
    'Season', 'Team', 'Player', 'gsis_id', 'Snaps','TeamSnaps', 'G', 'Rec', 'ReceivingYds', 'ReceivingTD', 'Tgt', 'air_yards', 'TargetP', 'TouchPerc', 'yards_after_catch'
]]\
.rename({
    'Season':'season',
    'Team':'team',
    'Player':'player_name',
    'Snaps':'snaps',
    'TeamSnaps':'team_snaps',
    'G':'games_played',
    'Rec':'receptions',
    'ReceivingYds':'rec_yards',
    'ReceivingTD':'rec_tds',
    'Tgt':'targets',
    'TargetP':'target_share',
    'TouchPerc':'touch_perc'
}, axis=1)\
.assign(
    stand_fp = lambda x: x.rec_yards*0.1 + x.rec_tds*6
)\
.assign(
    ppr_fp = lambda x: x.stand_fp + x.receptions
)\
.assign(
    half_ppr_fp = lambda x: x.stand_fp + x.receptions*0.5  
)\
.assign(
    adot = lambda x: x.air_yards / x.targets
).dropna()

print(wr_analysis_df.shape)
wr_analysis_df.head(25)
```

```python
def per_game_convert(column_name):
    wr_analysis_df[f'{column_name}_per_game'] = wr_analysis_df[column_name]/wr_analysis_df['games_played']

for col in ['snaps', 'targets', 'air_yards', 'yards_after_catch', 'rz_targets', 'rz_target_share', 'half_ppr_fp']:
    per_game_convert(col)

wr_analysis_df.head()
```

```python
lag_features = ['snaps', 'snaps_per_game', 'targets', 'targets_per_game', 'air_yards', 'air_yards_per_game', 'yards_after_catch','yards_after_catch_per_game',\
    'rz_targets', 'rz_targets_per_game', 'rz_target_share', 'half_ppr_fp', 'half_ppr_fp_per_game', 'games_played']
wr_analysis_df = wr_analysis_df.sort_values(by='season')

for lag in range(1, 9):
    shifted = wr_analysis_df.groupby('gsis_id').shift(lag)
    for col in lag_features:
        wr_analysis_df[f'lag_{col}_{lag}'] = shifted[col]

wr_analysis_df = wr_analysis_df.fillna(-1)

wr_analysis_df.head()
```

```python
pd.set_option('display.max_rows', None)
wr_analysis_df.corr()[['half_ppr_fp']].sort_values(by='half_ppr_fp', ascending=False)
```

```python
predict_df = wr_analysis_df.loc[wr_analysis_df.lag_games_played_1 >10].copy()
predict_df = predict_df.loc[predict_df.lag_snaps_1 >300]
predict_df = predict_df.dropna()
```

```python
features = ['lag_targets_1', 'lag_air_yards_1', 'lag_yards_after_catch_1', 'lag_yards_after_catch_1', 'lag_snaps_1', 'lag_rz_targets_1']
target = ['half_ppr_fp']

training_df = predict_df.loc[predict_df.season.isin(range(2013,2020))].copy()
training_df_X = training_df[features].values
training_df_y = training_df[target].values

test_df = predict_df.loc[predict_df.season == 2020]
test_df_X = test_df[features].values
test_df_y = test_df[target].values
```

```python
from sklearn.metrics import mean_absolute_error
lr = LinearRegression().fit(training_df_X, training_df_y)
y_pred =lr.predict(test_df_X)
mean_absolute_error(y_pred,test_df_y)
```

```python
final_2021_df = wr_analysis_df.loc[wr_analysis_df.season == 2021].copy().dropna()
final_2021_df['predicted_half_ppr'] = lr.predict(final_2021_df[features].values)
final_2021_df = final_2021_df[['player_name', 'predicted_half_ppr']]

pd.set_option('display.max_rows', None)
final_2021_df.sort_values(by='predicted_half_ppr', ascending=False).head(100)

# with pd.option_context('display.max_rows', None):
#     display(final_2021_df.sort_values(by='predicted_half_ppr', ascending=False)).head(100)
```
