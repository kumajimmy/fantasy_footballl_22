```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)

SNAPCOUNT_URL = 'https://raw.githubusercontent.com/fantasydatapros/data/master/snapcounts/{year}.csv'

start_year = 2013
end_year = 2021

snapcount_df = pd.DataFrame()

for year in range(start_year,end_year):
    yearly_sc_df = pd.read_csv(SNAPCOUNT_URL.format(year=year))
    yearly_sc_df['Season'] = year
    col_names = yearly_sc_df.columns.tolist()
    col_names.pop() #removes the last column name - 'Season'
    col_names.insert(0, 'Season')
    yearly_sc_df = yearly_sc_df[col_names]
    snapcount_df = pd.concat([snapcount_df, yearly_sc_df])

print(snapcount_df.shape)
snapcount_df.head()
```
